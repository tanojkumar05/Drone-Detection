```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the hand detection model
hand_model = YOLO('/content/drive/MyDrive/runs/detect/hand3.0_model/weights/best.pt')

# Path to the image
image_path = '/content/tello_52_jpeg.rf.21c816980ab0e3456cc6e96d7bbd8e0d.jpg'

# Run prediction
hand_result = hand_model.predict(source=image_path)

# Show the result
plt.imshow(hand_result[0].plot())
plt.title("Hand Detection")
plt.axis('off')
plt.show()
```
