```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load both models
hand_model = YOLO('/content/drive/MyDrive/runs/detect/hand_model/weights/best.pt')
drone_model = YOLO('/content/drive/MyDrive/runs/detect/drone_model/weights/best.pt')

# Path to the uploaded image
image_path = '/content/tello_52_jpeg.rf.21c816980ab0e3456cc6e96d7bbd8e0d.jpg'

# Run predictions
hand_result = hand_model.predict(source=image_path)
drone_result = drone_model.predict(source=image_path)

# Show results side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].imshow(hand_result[0].plot())
axes[0].set_title("Hand Detection")

axes[1].imshow(drone_result[0].plot())
axes[1].set_title("Drone Detection")

plt.show()
```
